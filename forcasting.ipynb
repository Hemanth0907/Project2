```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2016

```python
df_2016=pd.read_csv(r"2016 Stack Overflow Survey Responses.csv",encoding = 'ISO-8859-1')
```

```python
df_2016['Year']=2016
```

```python
df_2016.loc[df_2016['occupation'].str.contains('DevOps',na=False), 'occupation'] = 'DevOps'
df_2016['occupation'][df_2016['occupation'].str.contains('DevOps',na=False)]
df_2016['occupation'].value_counts()

category_counts = df_2016['occupation'].value_counts()

# Select the top 10 categories
top10_categories_2016 = category_counts.nlargest(20)

# Filter the data to include only the top 10 categories
data_2016_top10 = df_2016[df_2016['occupation'].isin(top10_categories_2016.index)]
cht=sns.catplot(data=data_2016_top10, x='occupation', kind="count", palette="ch:.25",height=5, aspect=2)

# Display the charts)
cht.set_xticklabels(rotation=90, fontsize=8)
plt.title("Top 20 Occupations in 2016")
df_2016_top10_country=df_2016[df_2016['occupation']=='DevOps']

category_counts = df_2016_top10_country['country'].value_counts()
# Select the top 10 categories
top10_categories_2016 = category_counts.nlargest(10)
# Filter the data to include only the top 10 categories
data_2016_top10_country = df_2016_top10_country[df_2016_top10_country['country'].isin(top10_categories_2016.index)]



cht=sns.catplot(data=data_2016_top10_country, x='country', kind="count",palette="ch:.25",height=5, aspect=2)
cht.set_xticklabels(rotation=90, fontsize=8)
plt.title("Top 10 country in 2016")
plt.show()


colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']
plt.pie(top10_categories_2016.values, labels=top10_categories_2016.index, colors=colors, autopct='%1.1f%%', startangle=90)

# Add a title to the chart
plt.title('Top 10 DevOps Occupations by Country in 2016')

# Show the chart
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

# define the percentage of DevOps respondents for the last six years
devops_percentage = [1.916830269498483, 7.8125, 9.659602448029943, 
                     9.970410539698255, 9.259862552551155, 8.458874147580868, 
                     8.421138832778293]

# define the years for the last six years
years = np.arange(2016, 2023)

# perform linear regression analysis
slope, intercept, r_value, p_value, std_err = linregress(years, devops_percentage)

# print the results
print("Slope: ", slope)
print("Intercept: ", intercept)
print("R-value: ", r_value)
print("P-value: ", p_value)
print("Standard error: ", std_err)
```

```python
sns.barplot(x=years, y=devops_percentage )

# set the title and axis labels
plt.title('Yearwise  Comparison of people worked on DevOps', fontsize=16)
plt.xlabel('year ', fontsize=12)
plt.ylabel('% of DevOps Respondents', fontsize=12)

# rotate the x-axis labels
plt.xticks(rotation=90)

# display the plot
plt.show()
```
